```python
# Q1. What is the total revenue generated by male vs. female customers?

# Group data by gender and calculate total revenue
gender_revenue = df.groupby('gender')['purchase_amount'].sum().reset_index()

# Plot
sns.barplot(data=gender_revenue, x='gender', y='purchase_amount', palette='coolwarm')

plt.title("Total Revenue Generated by Male vs Female Customers")
plt.xlabel("Gender")
plt.ylabel("Total Revenue")
plt.show()
```

Observation

male bar is higher  men share in total revenue dominate certain product categories or higher-value items.


```python
avg_purchase = df['purchase_amount'].mean()

# Filter customers who used a discount and spent above average
high_spenders_with_discount = df[(df['discount_applied'] == 'Yes') & (df['purchase_amount'] > avg_purchase)]

# Plot their spending compared to average
plt.figure(figsize=(8, 5))
sns.histplot(high_spenders_with_discount['purchase_amount'], bins=20, color='skyblue', kde=True)
plt.axvline(avg_purchase, color='red', linestyle='--', label=f'Average Purchase = {avg_purchase:.2f}')
plt.title("Customers Who Used Discounts But Spent Above Average")
plt.xlabel("Purchase Amount")
plt.ylabel("Count of Customers")
plt.legend()
plt.show()
```

A visible cluster far beyond the red average line indicates that discounts are not just attracting bargain hunters but some are still high paying customers.

```python
# Group by product and calculate average review rating
top_products = (
    df.groupby('item_purchased')['review_rating']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# Plot top 5 products by review rating
plt.figure(figsize=(8, 5))
sns.barplot(data=top_products, x='review_rating', y='item_purchased', palette='viridis')
plt.title("Top 5 Products with Highest Average Review Rating")
plt.xlabel("Average Review Rating")
plt.ylabel("Product Name")
plt.show()

# Display the top 5 products data
top_products
```

These products likely have high satisfaction and retention potential, and can be promoted as hero products in marketing campaigns.
Since theres a narrow gap between ratings, it suggests consistent product quality.

```python
# Calculate average purchase amount per shipping type
avg_shipping_spend = (
    df.groupby('shipping_type')['purchase_amount']
    .mean()
    .reset_index()
    .sort_values(by='purchase_amount', ascending=False)
)

# Plot
plt.figure(figsize=(7, 5))
sns.barplot(data=avg_shipping_spend, x='shipping_type', y='purchase_amount', palette='coolwarm')
plt.title("Average Purchase Amount by Shipping Type")
plt.xlabel("Shipping Type")
plt.ylabel("Average Purchase Amount")
plt.show()
```

2-Day Shipping and Express shipping shows higher spend, it suggests these customers are premium buyers, worth targeting for loyalty or faster service perks.

```python
# Q5. Do subscribed customers spend more? Compare average spend and total revenue between subscribers and non-subscribers.
fig, axes = plt.subplots(1, 2, figsize=(9, 5))

sns.barplot(data=df,x='subscription_status',y='purchase_amount',estimator='mean',ax=axes[0])
axes[0].set_title("Average Spend")
axes[0].set_xlabel("Subscription Status")
axes[0].set_ylabel("Average Purchase Amount")

sns.barplot(data=df,x='subscription_status',y='purchase_amount',estimator=sum,ax=axes[1])
axes[1].set_title("Total Revenue")
axes[1].set_xlabel("Subscription Status")
axes[1].set_ylabel("Total Revenue")

plt.suptitle("Subscribed vs Non-Subscribed  Spending Comparison", fontsize=14)
plt.tight_layout()
plt.show()
```

Average amount of subscribers and non subscribers is almost same. But the revenue generated by non subscribers are huge, That means non subscribers are spending more many as in total but the average amount is almost same.

```python
# Q6. Which 5 products have the highest percentage of purchases with discounts applied?
df['discount_applied'] = df['discount_applied'].astype(str).str.strip().str.lower()

# Replace common yes/no formats with numeric
df['discount_applied'] = df['discount_applied'].replace({
    'yes': 1,
    'no': 0,
    'y': 1,
    'n': 0,
    'true': 1,
    'false': 0
})

# Convert to numeric (any invalid value  NaN  replace with 0)
df['discount_applied'] = pd.to_numeric(df['discount_applied'], errors='coerce').fillna(0)

# ------------------ CALCULATION STEP ------------------

discount_rate = (
    df.groupby('item_purchased')['discount_applied']
      .mean()
      .sort_values(ascending=False)
      .head(5)
      .reset_index()
)

plt.figure(figsize=(10, 5))
sns.barplot(data=discount_rate, x='item_purchased', y='discount_applied')
plt.title("Top 5 Items with Highest Discount Usage (%)")
plt.xlabel("Item Purchased")
plt.ylabel("Discount Usage Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q7. Segment customers into New, Returning, and Loyal based on their total number of previous purchases, and show the count of each segment.
def segment(p):
    if p == 1:
        return 'New'
    elif p <= 10:
        return 'Returning'
    else:
        return 'Loyal'

df['customer_segment'] = df['previous_purchases'].apply(segment)

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='customer_segment')
plt.title("Customer Segments")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.show()
```

The Number of loyal customers are is huge that means many people are liking the product hence they are reapeating their orders

```python
# Q8. What are the top 3 most purchased products within each category?
top3 = (
    df.groupby(['category', 'item_purchased'])
    .size()
    .reset_index(name='purchase_count')
    .sort_values(['category', 'purchase_count'], ascending=[True, False])
    .groupby('category')
    .head(3)
)

plt.figure(figsize=(10,5))
sns.barplot(data=top3, x='item_purchased', y='purchase_count', hue='category')
plt.title("Top 3 Most Purchased Items in Each Category")
plt.xlabel("Item Purchased")
plt.ylabel("Purchase Count")
plt.xticks(rotation=45)
plt.show()
```

items in Clothing category have greater number of purchase count indicating people are more likely to buy cloths

```python
# Q9. Are customers who are repeat buyers (more than 5 previous purchases) also likely to subscribe?
# Clean subscription_status
df['subscription_status'] = df['subscription_status'].astype(str).str.strip().str.lower()

df['subscription_status'] = df['subscription_status'].replace({
    'yes': 1,
    'no': 0,
})

df['subscription_status'] = pd.to_numeric(df['subscription_status'], errors='coerce').fillna(0)
df['repeat_buyer'] = df['previous_purchases'] > 11

# Group and calculate subscription rate
repeat_sub = (
    df.groupby('repeat_buyer')['subscription_status']
    .mean()
    .reset_index()
)
plt.figure(figsize=(6,4))
sns.barplot(data=repeat_sub, x='repeat_buyer', y='subscription_status')
plt.title("Subscription Rate: Repeat vs Non-Repeat Buyers")
plt.xlabel("Repeat Buyer (>5 Previous Purchases)")
plt.ylabel("Subscription Rate")
plt.show()
```

```python
# Q10. What is the Total Customer count in each age group?
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='age_group')
plt.title("Revenue Contribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Customers")
plt.show()
```

Young adults are the most in numbers means they hold the majority 

```python
# Q11. What is the revenue contribution of each age group?
plt.figure(figsize=(7,4))
sns.barplot(data=df, x='age_group', y='purchase_amount', estimator=sum)
plt.title("Revenue Contribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Revenue")
plt.show()
```

Most revenue is generated by the young adults. As the are the most spenders as well as they are more in number.
